## Question 1 a

```python
def survival_prob_func(B):
    d2 = (np.log(15000000/B) + (0.012 - 0.21^2/2)*5 )/(0.21 * np.sqrt(5))
    s = 1 - norm.cdf(-d2)
    return s
```

```python
B = np.arange(1000000, 50000000, 1000)
plt.plot(B,survival_prob_func(B))
plt.xlabel('B face value')
plt.ylabel('Probability')
plt.title('Survival Probability')
```
